You can control the aesthetics of your graphs with seaborn, by default it'll use the matplotlib theme, however, making your graphs look attractive is important.

Here are a few examples of adjusting the theme / style of our graphs.

```python
import numpy as np
import matplotlib.pyplot as plt

# A simple sin graph.
#sns.set_style('whitegrid')
#sns.set_style('white')
#sns.set_style('ticks')
sns.set_context('notebook')

x = np.linspace(0, 20, 150)
for i in range(1, 11):
    plt.plot(x, np.sin(x + i * .5) * (10 + 2 - i) * 1)
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
# Create visualisation
sns.relplot(data=df,
            x='total_bill', y='tip', col='time',
            hue='smoker', style='smoker', size='size')
```

Distributional Representation

```python
sns.displot(data=df,
            x='total_bill', col='time')
```

```python
sns.displot(data=df, kind='ecdf',
            x='total_bill', col='time')
```

Categorical Plots

```python
sns.catplot(data=df, kind='swarm',
            x='day', y='total_bill', hue='smoker')
```

```python
sns.catplot(data=df, kind='violin',
            x='day', y='total_bill',
            hue='smoker', split=True)
```
